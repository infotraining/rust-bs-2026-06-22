# Vector

```rust
fn tokenize(s: &str, separator: char) -> Vec<&str> {
    let mut tokens = Vec::new();
    let mut start = 0;

    for (i, c) in s.char_indices() {
        if c == separator {
            tokens.push(&s[start..i]);
            start = i + c.len_utf8();
        }
    }

    tokens.push(&s[start..]);

    tokens
}

fn main() {
    let s = "hello,world,this is a test,!";
    let tokens = tokenize(s, ',');
    assert_eq!(tokens, vec!["hello", "world", "this is a test", "!"]);

    println!("Success!");
}

main(); 
```

```rust
fn split_in_two(v: Vec<i32>) -> (Vec<i32>, Vec<i32>) {
    let mid = v.len() / 2;
    let first_half = v[..mid].to_vec();
    let second_half = v[mid..].to_vec();
    (first_half, second_half)
}

fn main() {
    let v = vec![1, 2, 3, 4, 5, 6];
    let (first_half, second_half) = split_in_two(v);
    assert_eq!(first_half, vec![1, 2, 3]);
    assert_eq!(second_half, vec![4, 5, 6]);

    println!("Success!");
}

main();
```
