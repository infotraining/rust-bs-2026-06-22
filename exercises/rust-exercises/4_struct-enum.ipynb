# Enum

1.

```rust
// Fix the errors
enum Number {
    Zero,
    One,
    Two,
}

#[derive(PartialEq, PartialOrd)]
enum Number1 {
    Zero = 0,
    One,
    Two,
}

// C-like enum
enum Number2 {
    Zero = 0,
    One = 1,
    Two = 2,
}


fn main() {
    // An enum variant can be converted to a integer by `as`
    assert_eq!(Number::One as i32, Number1::One as i32);
    assert_eq!(Number1::One as i32, Number2::One as i32);
    assert!(Number1::One < Number1::Two); 

    println!("Success!");
} 

main();
```

# Impl

1. Implement functions for `struct`

```rust
struct ContactEntry(String, String);

impl ContactEntry {
    fn new(name: &str, phone: &str) -> ContactEntry {
        ContactEntry(name.to_string(), phone.to_string())
    }

    fn name<'a>(&'a self) -> &'a str {
        &self.0
    }

    fn phone(&self) -> &str {
        &self.1
    }
}

fn main() {
    let c1 = ContactEntry::new("John", "666-66-66-66");
    
    assert_eq!(c1.name(), "John");
    assert_eq!(c1.phone(), "666-66-66-66");    
}
```
